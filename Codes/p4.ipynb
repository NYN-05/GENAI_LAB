```python
import gensim.downloader as api
from transformers import pipeline
```

```python
model=api.load("glove-wiki-gigaword-100")
print("Model loaded successfully")
```

```python
generator= pipeline('text-generation', model="gpt2")
```

```python
prompt1 = "Explain elections"
```

```python
result1= generator( prompt1 ,max_length=30, truncation = True)
print(result1)
```

```python
similar=model.most_similar("vote",topn=10)
for word in similar:
  print(word[0])
```

```python
store = [str(word) for word, _ in similar]
prompt2 = prompt1 + " including " + ", ".join(store)
```

```python
result2 = generator(prompt2, max_new_tokens=30,truncation=True)
print(result2)
```

```python
import gensim.downloader as api
from transformers import pipeline
m = api.load("glove-wiki-gigaword-100")
g = pipeline("text-generation", model="gpt2")
p = "Explain cricket"
print("Original Prompt Output:\n")
print(g(p, max_length=50)[0]['generated_text'])
sim = m.most_similar("cricket", topn=5)
print("Similar words:\n")
for w,s in sim: print(w, s)
ep = p + " including " + ", ".join([w for w,_ in sim])
print("\nEnriched Prompt:\n", ep)
print("\nEnriched Prompt Output:\n")
print(g(ep, max_length=80)[0]['generated_text'])
```
