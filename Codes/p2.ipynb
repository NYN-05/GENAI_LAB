## Use dimensionality reduction (e.g., PCA or t-SNE) to visualize word embeddings for Q 1. Select 10 words from a specific domain (e.g., sports, technology) and visualize their embeddings. Analyze clusters and relationships. Generate contextually rich outputs using embeddings. Write a program to generate 5 semantically similar words for a given input.

```python
import gensim.downloader as api
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
model = api.load("glove-wiki-gigaword-100")
print("Model loaded successfully!")
```

```python
words = ["cricket", "football", "tennis", "bat", "ball", "stadium","player","coach","match","tournament"]
```

```python
word_vectors = []
for word in words:
    word_vectors.append(model[word])
word_vectors = np.array(word_vectors)
```

```python
pca=PCA(n_components=2)
reduced_vectors=pca.fit_transform(word_vectors)
```

```python
plt.figure(figsize=(8,6))

for i, word in enumerate(words):
    x=reduced_vectors[i, 0]
    y=reduced_vectors[i, 1]
    plt.scatter(x, y)
    plt.text(x+0.01, y+0.01, word)
plt.title("Word Embeddings Visualized (Sports Domain)")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.show()
```

```python
word =input("Enter a word to find similar words: ")
similar_words = model.most_similar(word, topn=5)
print(f"Top 5 similar words to '{word}':")
for ww, score in similar_words:
    print(f"{ww}: {score:.4f}")
```

```python
# pip install gensim matplotlib scikit-learn

import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load pre-trained model
model = api.load("word2vec-google-news-300")

# Words list
words = ['computer','internet','software','hardware','keyboard',
         'mouse','server','network','programming','database']

# Get vectors and reduce dimensions
reduced = PCA(n_components=2).fit_transform([model[w] for w in words])

# Similar words
print("Top 5 words similar to 'computer':")
for w, s in model.most_similar('computer', topn=5):
    print(f"{w}: {s:.4f}")

# Plot
for i, w in enumerate(words):
    plt.scatter(*reduced[i])
    plt.text(*reduced[i], w)

plt.title("Word Embeddings (PCA)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.grid()
plt.show()
```
