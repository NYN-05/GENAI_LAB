```python
# Minimal: Pre-trained Word Vectors + Relationships + Arithmetic

# !pip install gensim
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity

# Load model
m = api.load('glove-twitter-25')

# Relationships
print(m.most_similar('king', topn=3))

# Vector arithmetic
print(m.most_similar(positive=['king','woman'], negative=['man'], topn=3))  # king - man + woman
print(m.most_similar(positive=['paris','italy'], negative=['france'], topn=3))  # paris - france + italy

# Similarity analysis
print(cosine_similarity(m['king'].reshape(1,-1), m['queen'].reshape(1,-1))[0][0])
```
